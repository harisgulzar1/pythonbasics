```python
import pandas as pd
import matplotlib.pyplot as plt


# df = pd.read_csv("sample_data/california_housing_test.csv")
df = pd.read_csv("sample_data/sampleData1.csv")


plt.figure(figsize=(12,12), dpi=120)

# plt.scatter(df["housing_median_age"], df["median_house_value"], linewidth=1)
plt.scatter(df["X"], df["Y"], linewidth=1)
```

```python
from sklearn.cluster import KMeans
# importing function from the sub library

clusterNum = 6

# df = df.query("total_rooms>3000 & total_rooms<3500")
model = KMeans(n_clusters=clusterNum, init='k-means++')
res = model.fit_predict(df)

df["cluster"] = res
df

fig = plt.figure(figsize=(12,12),dpi=120)
for i in range(clusterNum):
    plt.scatter(df[df["cluster"]==i]["X"],df[df["cluster"]==i]["Y"],label="c"+str(i))
plt.legend()
```

```python
from sklearn.cluster import KMeans
# importing function from the sub library

clusterNum = 6

# df = df.query("total_rooms>3000 & total_rooms<3500")
model = KMeans(n_clusters=clusterNum, init='k-means++')
res = model.fit_predict(df)
centroids = model.cluster_centers_

df["cluster"] = res
df



fig = plt.figure(figsize=(4,4),dpi=120)
for i in range(clusterNum):
    plt.scatter(df[df["cluster"]==i]["X"],df[df["cluster"]==i]["Y"],label="c"+str(i))
plt.legend()

plt.scatter(centroids[:,0] , centroids[:,1] , s = 20, color = 'k', marker='x')
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

clusterNumList = []
scoreList = []

# df = df[["housing_median_age", "median_house_value"]]

for i in range(2,11):
    model = KMeans(n_clusters=i, init='k-means++')
    res = model.fit_predict(df)
    score = silhouette_score(df, res, metric='euclidean')
    print(i,score)
    clusterNumList.append(i)
    scoreList.append(score)

fig = plt.figure(figsize=(8,8))
plt.plot(clusterNumList,scoreList,marker="o")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler
import plotly

def plot_dendrogram(model, **kwargs):

    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack([model.children_, model.distances_,
                                      counts]).astype(float)

    dendrogram(linkage_matrix, **kwargs)

    return linkage_matrix


df = pd.read_csv("sample_data/lesson2-exercise-data1.csv")
#c = StandardScaler()
#f = pd.DataFrame(sc.fit_transform(df))
#f.columns = ["X","Y","Z"]

model = AgglomerativeClustering(affinity='euclidean', linkage='ward',distance_threshold=0, n_clusters=None)

model = model.fit(df)
fig = plt.figure(figsize=(8,8),dpi=150)
linkage_matrix = plot_dendrogram(model, truncate_mode='level', p=4)
plt.xlabel("Number of points in node (or index of point if no parenthesis).")
plt.show()


#シルエット係数によるクラスタ数の選択
#クラスタ内は密に集まっているほど良く、クラスタ間は遠く離れているほど良いことを表す指標
#シルエット係数は、[-1,1]。大きいほど、良いことを表す。



clusterNumList = []
scoreList = []

for i in range(2,11):
    clusters = fcluster(linkage_matrix, t=i, criterion='maxclust')
    score = silhouette_score(df, clusters, metric='euclidean')
    clusterNumList.append(i)
    scoreList.append(score)


    fig = plt.figure(figsize=(5,5),dpi=120)
    df["cluster"] = clusters

    ax = Axes3D(fig)

    ax.set_title(str(i) + " Clusters")
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
    for j in range(1,i+1):
        ax.plot(df[df["cluster"]==j].X,df[df["cluster"]==j].Y,df[df["cluster"]==j].Z,marker="o",linestyle='None',label="c"+str(j))

    plt.show()


fig = plt.figure(figsize=(8,8))
plt.plot(clusterNumList,scoreList,marker="o")
```
