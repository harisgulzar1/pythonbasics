**Find the distance between two points!**

```python
!pip install haversine # a library to get geographical distance (Because it's not euclidean distance)
```

```python
# A function to convert Series coordinate into absolute coordinates

def coordinates(centroid):
    import numpy as np
    lat = np.array(centroid.y)[0] # converting GeoSories back into normal data
    lon = np.array(centroid.x)[0]
    return (lat, lon) # a pair of latitude and longitude
```

```python
#Draw a line between Japan and Qatar

import geopandas as gpd
import geopy.distance as gdis
import numpy as np

import haversine as gdis

japan_df = global_df.query("name=='Japan'")
japan_center = japan_df.centroid
japan_coordinates = coordinates(japan_center)

print(japan_coordinates)

qatar_df = global_df.query("name=='Qatar'")
qatar_center = qatar_df.centroid
qatar_coordinates = coordinates(qatar_center)

print(qatar_coordinates)

geo_distance = gdis.haversine(japan_coordinates, qatar_coordinates)

print("The distance between Japan and Qatar is %0.2f" % geo_distance, "km.")
```

```python
def coordinates(centroid):
    import numpy as np
    lat = np.array(centroid.y)[0] # converting GeoSories back into normal data
    lon = np.array(centroid.x)[0]
    return (lon, lat) # longitude and latitude
```

```python
# Plot line b/w Qatar and Japan

import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import LineString

japan_df = global_df.query("name=='Japan'")
japan_center = japan_df.centroid
japan_coordinates = coordinates(japan_center)


qatar_df = global_df.query("name=='Qatar'")
qatar_center = qatar_df.centroid
qatar_coordinates = coordinates(qatar_center)


line = LineString([japan_coordinates, qatar_coordinates])
line_df = gpd.GeoDataFrame(geometry=[line])

line_df # Line in DataFrame Format

# plot continet
continet = continent_df.plot(figsize=(10,10))

# add ine
line_df.plot(ax=continet, color='red')
```
